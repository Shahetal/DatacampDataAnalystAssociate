```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataframe
df = pd.read_csv(r"C:\Users\shaha\Downloads\pet_supplies_2212.csv")
df.head(10)
```

```python
#checking data info
df.info()
```

```python
#looking out for duplicate entries
df.duplicated().sum()
```

```python
#replacing '-' with 'Unknown'
df['category'] = df['category'].replace('-','Unknown')
df.info()
df.head(10)
```

```python
#replacing values to the proper case
df['size']=df['size']. apply(str.capitalize)
df.head(10)
```

```python
#looking through prices colum to see the individual values in it
df['price'].value_counts()
```

```python
#replacing wrong entries in the prices column with NaN
df['price'].replace((['unlisted'], [np.NaN]), inplace =True)
```

```python
#filling NaN values with the median value of the prices column as instructed
df['price'] = df['price'].fillna(df['price'].median(), inplace= False)
```

```python
#converting datatypes of the prices column
df['price'] = df['price'].astype('float')
df.head(10)
```

```python
#converting datatypes of the sales column
df['sales'] = df['sales'].astype('float')
df.head(10)
```

```python
#filling up null values in the rating column with 0 as instructed
df['rating'] = df['rating'].fillna('0', inplace = False)
df.head(25)
```

```python
#changing datatype for rating column
df['rating'] = df['rating'].astype(int)
df.head(10)
```

```python
#This code will create a separate donut chart for each animal in your data, with the total sales for each category in the outer ring and the percentage of repeat purchases for each category in the inner ring. The code uses a for loop to iterate over each unique animal in your data and creates a donut chart for each one.
# group the data by animal and category, and calculate the sums
sums = df.groupby(['animal', 'category']).sum()

# iterate over each animal and create a donut chart
for animal in df['animal'].unique():
    animal_df = sums.loc[animal]
    total_sales = animal_df['sales'].sum()
    total_repeat = animal_df['repeat_purchase'].sum()
    
    # create a pie chart with the total sales for the current animal
    fig, ax = plt.subplots()
    ax.pie(animal_df['sales'], labels=animal_df.index.get_level_values('category'), autopct='%1.1f%%')

    # add a smaller pie chart in the center with the percentage of repeat purchases for the current animal
    center_circle = plt.Circle((0,0), 0.5, color='white')
    ax.add_artist(center_circle)
    ax.pie(animal_df['repeat_purchase'], colors=['yellow', 'green', 'red', 'blue'], radius=0.5, wedgeprops=dict(width=0.2, edgecolor='white'), labels=animal_df.index.get_level_values('category'), labeldistance=0.4)

    # add a title
    ax.set_title(f'{animal} Sales and Repeat Purchases')

    # show the chart
    plt.show()
```
